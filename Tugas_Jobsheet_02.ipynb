```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

data1 = pd.read_csv('50_Startups.csv')
data1.head()
```

```python
transformer_list = [
     ('encoded', OneHotEncoder(dtype='int'), ['State']),
     ('skip', 'passthrough', ["R&D Spend", "Administration", "Marketing Spend", "Profit"])
]

coltran = ColumnTransformer(transformer_list)
tran = coltran.fit_transform(data1)

data2 = pd.DataFrame(tran, columns=coltran.get_feature_names_out())
data2.head()
```

```python
d = data2.iloc[:, :-1].values
i = data2.iloc[:, 6].values

d = i.reshape(len(i), 1)
i.shape
```

```python
dm_train, dm_test, im_train, im_test = train_test_split(d, i, test_size=0.2, random_state=50)
#linear regression
mlr = LinearRegression()
mlr.fit(dm_train, im_train)

im_pred = mlr.predict(dm_test)

mix = np.concatenate((im_test, im_pred), axis=1)
mix
```

```python
r2_mlr = r2_score(im_test, im_pred)
print(r2_mlr)
```
